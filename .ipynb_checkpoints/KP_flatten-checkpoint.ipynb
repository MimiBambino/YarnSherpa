```python
import json
with open("yarnsherpa-export.json") as json_file:
    data = json.load(json_file)   
len(data)
```

```python
len(data[0]["colorways"])
```

```python
new_list = []
for i in range(len(data)):
    for j in range(len(data[i]["colorways"])):
        new_dict = {}
        new_dict["manufacturer"] = data[i]["manufacturer"]
        new_dict["line"] = data[i]["line"]
        new_dict["link"] = data[i]["link"]
        new_dict["name"] = data[i]["colorways"][j]["name"]
        new_dict["color"] = data[i]["colorways"][j]["color"]
        new_dict["weight"] = data[i]["weight"]
        new_dict["fiber"] = data[i]["fiber"]
        new_dict["image"] = data[i]["colorways"][j]["image"]
        new_list.append(new_dict)
new_list
```

```python
len(new_list)
```

```python
from firebase import Firebase
f = Firebase('https://yarnsherpa.firebaseio.com/yarns')
r = f.put(new_list)
```
